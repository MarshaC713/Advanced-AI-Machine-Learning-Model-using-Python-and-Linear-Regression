Analysis of the values of the different quartiles to find the IQR.
The IQR (interquartile range) is the difference between the 75th (q3)
and 25th (q1) percentile of the data.
It is a measure of the dispersion similar to standard deviation or variance,
but is much more robust against outliers.
to do this I first confirm and determine the values of each:

```python
# The 75th(q3) and 25th(q1) quartiles are 85.82, and 55.29 respectively for the Power_perf_factor
# as returned by the describe() function

q3 = df['Power_perf_factor'].quantile(0.75)
q1 = df['Power_perf_factor'].quantile(0.25)
print(q3)
print(q1)
```

```python
# Finding the IQR (interquartile range):

IQR = q3 - q1
print(IQR)
```

A commonly used rule says that a data point is an outlier if it is more than 1.5 * IQR above the third quartile or below the first quartile.
In other words low outliers are below q1 - (IQR * 1.5).
High outliers are above q3 + (IQR * 1.5)

```python
# Finding the IQR (interquartile range)
# values below 9.5 are outliers:
lower_limit = q1 - (IQR * 1.5)
print(lower_limit)
```

```python
# values above 131.625 are outliers:

upper_limit = q3 + (IQR * 1.5)
print(upper_limit)
```

```python
# Detecting the outliers for removal:
df_outliers = np.where(df['Power_perf_factor'] > upper_limit, True,
                       np.where(df['Power_perf_factor'] < lower_limit, True, False))
```

```python
# Excluding the outliers to confirm mean (average) value:
df_without_outliers = df.loc[~(df_outliers)]
```

```python
# Removing the outliers to confirm and display the average Power_perf_factor
# in a box plot diagram:

a=sns.boxplot( y='Power_perf_factor', data=df_without_outliers)
a
```

#### AI and Machine Learning Model

```python
plt.scatter(df['Power_perf_factor'], df['Price_in_thousands']) 
plt.title('Power_perf_factor vs Price_in_thousands')  
plt.xlabel('Power_perf_factor')  
plt.ylabel('Price_in_thousands')  
plt.show()
```

```python
# Getting the dependent and independent variables and 
# reshaping them using pandas and numpy, so that I can use test_train_split.
# Pandas and numpy support multidimensional arrays such as those produced below:

X = df['Power_perf_factor'].values.reshape(-1,1)
y = df['Price_in_thousands'].values.reshape(-1,1) 
```

```python
#Splitting data into training sets and testing sets
# By Marsha Castello

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
y_test
```

```python
#Training my Model
model = LinearRegression().fit(X_train, y_train)
```

```python
X_train, y_train
```

```python
#Seeing how accurate my model is, with a score of 1 being the best 
#possible score, therefore 0.77 is a good model.

print("Training set score:", model.score(X_train, y_train))
print("Test set score:", model.score(X_test, y_test))
```

```python
#Making a prediction for our test data.
y_prediction = model.predict(X_test)
y_prediction
```

```python
#Plotting the results and line of best fit using matplotlib:
plt.scatter(X_test, y_test)
plt.plot(X_test, y_prediction, color='r')
plt.title('Power_perf_factor vs Price_in_thousands')  
plt.xlabel('Power_perf_factor')  
plt.ylabel('Price_in_thousands')  
plt.show()
```
